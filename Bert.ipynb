# **What is BERT?**

- BERT is an Artificial Intelligence model developed by Google that helps computers understand human language more like humans do.
                                 

In simple terms,
- BERT reads a sentence and understands the meaning of words based on the full context, not just individually.
                                                                                                                                                                            
# **Why was BERT needed?**

- **Earlier language models had a limitation:**

- They read text only in one direction
(either left to right or right to left)
Because of this, they often misunderstood the meaning of words

**Example:**

**Sentence:**
- "I saw a bat"

The word "bat" can mean:

- an animal
- a sports tool

Older models could not clearly understand which meaning is correct.

# **What makes BERT different?**

- BERT reads the sentence in both directions at the same time.
This is called bidirectional understanding.

It looks at:

words before a term
words after a term

Then it decides the correct meaning.

**Example:**

"He is playing cricket with a bat"

BERT understands that "bat" refers to a sports tool because of the surrounding words.

# **How does BERT work?**                                   
- **Input:** You give a sentence
- **Tokenization:** The sentence is broken into smaller parts (words or sub-words)
- **Context** Understanding: BERT analyzes how each word relates to others
- **Output:** It gives a result such as sentiment, emotion, or classification  

# **What is inside BERT?**           
                                       

- BERT is built on something called the
Transformer Architecture


**The key idea inside it is self-attention:**                        

- Every word looks at every other word in the sentence
This helps it understand relationships and meaning better
Real-world applications of BERT

**BERT is widely used in:**

- Sentiment analysis (positive or negative text)           
- Chatbots and virtual assistants          
- Search engines (improving search results)           
- Spam detection                   
- Mental health text analysis (like your project)                                     


# **Why is BERT powerful? **
                   
- It understands context, not just words             
- It works well on complex sentences               
- It produces more accurate results than older models

# **2. Import Libraries**

```python
from transformers import pipeline
```

# **3. Initialize Model**

```python
# Mental health / depression classifier
classifier = pipeline(
    "text-classification",
    model="bhadresh-savani/distilbert-base-uncased-emotion"
)
```

# **4. Define Risk Level Function**

```python
def get_risk_level(label):
    if label in ["sadness", "fear"]:
        return "HIGH"
    elif label in ["anger"]:
        return "MEDIUM"
    else:
        return "LOW"
```

# **5. Define Suggestion Function**

```python
def get_suggestion(risk):
    if risk == "HIGH":
        return "Consider talking to someone you trust or a professional."
    elif risk == "MEDIUM":
        return "Try relaxation techniques and take short breaks."
    else:
        return "You seem fine. Keep maintaining a positive routine."
```

# **6. Main App Function**

```python
def run_app():
    print("--- Mental Health Detection App (BERT) ---")
    print("Type 'quit' to exit.")

    while True:
        user_input = input("\nEnter text: ")

        if user_input.lower() == 'quit':
            break

        # Model Prediction
        result = classifier(user_input)[0]

        label = result['label']
        score = result['score']

        # Risk + Suggestion
        risk = get_risk_level(label)
        suggestion = get_suggestion(risk)

        # Output
        print(f"\nEmotion: {label}")
        print(f"Confidence: {score:.2%}")
        print(f"Risk Level: {risk}")
        print(f"Suggestion: {suggestion}")
```

# **7. Run the App**

```python
run_app()
```
